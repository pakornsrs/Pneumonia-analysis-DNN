```python
import cv2
import h5py
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Access to image's folder in local storage

# For macOs
# path = glob.glob("\your folder path\*.jpg") 
# For windows
path = glob.glob(r"C:\your folder path\*.jpeg")          

images = [img for img in path]  
images.sort()                   
img_x_train = []                

for img in images:
    img_read = cv2.imread(img)                             
    img_resize = cv2.resize(img_read,(128,128))               
    img_RGB = cv2.cvtColor(img_resize,cv2.COLOR_BGR2RGB)    
    img_x_train.append(img_RGB)                            
```
